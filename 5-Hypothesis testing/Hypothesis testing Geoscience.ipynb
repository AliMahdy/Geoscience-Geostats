```python
import numpy as np
from scipy import stats

# Reproducible synthetic data (mD = millidarcies)
np.random.seed(101)
formation_A = np.random.normal(loc=150, scale=20, size=30)
formation_B = np.random.normal(loc=180, scale=25, size=30)

# Print sample stats
print(f"Formation A Mean: {np.mean(formation_A):.2f} mD")
print(f"Formation B Mean: {np.mean(formation_B):.2f} mD\n")

# Perform t-test
t_stat, p_val = stats.ttest_ind(formation_A, formation_B)
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_val:.4e}")

# Decision rule (α = 0.05)
alpha = 0.05
if p_val < alpha:
    print("\n✅ Reject H₀: There is a statistically significant difference in permeability.")
else:
    print("\n❌ Fail to reject H₀: No significant difference detected.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid", font_scale=1.1)
plt.figure(figsize=(9, 5))

# Plot distributions
sns.kdeplot(formation_A, color="steelblue", fill=True, label="Formation A", alpha=0.6)
sns.kdeplot(formation_B, color="tomato", fill=True, label="Formation B", alpha=0.6)

# Add mean lines
plt.axvline(np.mean(formation_A), color="navy", linestyle="--", linewidth=1.8)
plt.axvline(np.mean(formation_B), color="darkred", linestyle="--", linewidth=1.8)

# Labels & styling
plt.title("Permeability Comparison: Formation A vs. Formation B", fontsize=14, pad=15)
plt.xlabel("Permeability (mD)", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.legend(title="Rock Unit", frameon=True)
plt.tight_layout()
plt.savefig("post_5_illustration.png", dpi=300, bbox_inches='tight')
plt.show()
```
