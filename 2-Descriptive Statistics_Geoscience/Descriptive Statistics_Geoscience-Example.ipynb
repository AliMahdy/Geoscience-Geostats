```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd
```

```python
# Read data from CSV
df = pd.read_csv('synthetic_perm_data.csv')
perm = df['perm']
```

```python
# The essential statistics
mean_perm = np.mean(perm)
median_perm = np.median(perm)
std_perm = np.std(perm)
skewness = stats.skew(perm)
kurtosis = stats.kurtosis(perm)
p90_p10 = np.percentile(perm, 90) / np.percentile(perm, 10)

print(f"Mean: {mean_perm:.1f} mD")
print(f"Median: {median_perm:.1f} mD")
print(f"Skewness: {skewness:.2f}")
print(f"P90/P10 ratio: {p90_p10:.1f}")
```

```python
# Visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
ax1.hist(perm, bins=30, alpha=0.7, color='skyblue')
ax1.axvline(mean_perm, color='red', linestyle='--', label=f'Mean: {mean_perm:.1f}')
ax1.axvline(median_perm, color='green', linestyle='--', label=f'Median: {median_perm:.1f}')
ax1.set_xlabel('Permeability (mD)')
ax1.legend()

ax2.boxplot(perm)
ax2.set_ylabel('Permeability (mD)')
plt.show()
```
